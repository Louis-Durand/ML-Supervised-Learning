## Part 1 : artificial dataset generation

```python
import numpy as np
import pandas as pd
```

Generation of 2 columns not correlated

```python
column_0 = np.round(20 * np.absolute(np.random.random(300))+ 30, 2)# floats from 30 to 50
column_1 = np.round(3 * np.absolute(np.random.random(300)) + 1, 2)# floats from 1 to 4

df = pd.DataFrame(column_0)
df.insert(len(df.columns), len(df.columns), column_1)
```

Generation of 2 columns closely positively correlated 

```python
positive_correlated_col_2 = range(100,700,2)
positive_correlated_col_3 = range(0,900, 3)
df.insert(len(df.columns), len(df.columns), positive_correlated_col_2)
df.insert(len(df.columns), len(df.columns), positive_correlated_col_3)
```

Generation of a column closely negatively correlated with column 3

```python
negative_correlated_col_4 = np.arange(850.0,400.0,-1.5)
df.insert(len(df.columns), len(df.columns), negative_correlated_col_4)
```

Generation of a sixth column

```python
col_6 = range(0,-300,-1)
df.insert(len(df.columns), len(df.columns), col_6)
```
